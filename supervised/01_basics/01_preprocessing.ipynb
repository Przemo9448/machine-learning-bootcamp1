```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit_transform(df['bought'])
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
df['bought'] = le.inverse_transform(df['bought'])
df
```

# OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, drop='first')
encoder.fit_transform(df[['size']])
```

```python
encoder.categories_
```

# get_dummies()

```python
pd.get_dummies(data=df, dtype=int)
```

```python
pd.get_dummies(data=df, dtype=int, drop_first=True)
```

# StandardScaler

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
def standardscaler(data):
  return (data - data.mean()) / data.std()

standardscaler(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```

```python
df = df_raw.copy()

le = LabelEncoder()
df['bought'] = le.fit_transform(df['bought'])

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df = pd.get_dummies(data=df, dtype=int, drop_first=True)
df
```
